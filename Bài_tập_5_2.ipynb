```python
import random
import math
from collections import Counter

def is_prime(n):
  if n < 2:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True

def find_decreasing_subsequences(data):
  if not data:
    return []

  subsequences = []
  current_subsequence = [data[0]]

  for i in range(1, len(data)):
    if data[i] < current_subsequence[-1]:
      current_subsequence.append(data[i])
    else:
      if len(current_subsequence) > 1:
        subsequences.append(current_subsequence)
      current_subsequence = [data[i]]

  if len(current_subsequence) > 1:
    subsequences.append(current_subsequence)

  return subsequences

if __name__ == "__main__":
  while True:
    try:
      n_str = input("Nhập số lượng phần tử N: ")
      n = int(n_str)
      if n <= 0:
        print("N phải là số nguyên dương. Vui lòng nhập lại.")
      else:
        break
    except ValueError:
      print("Giá trị nhập không hợp lệ. Vui lòng nhập một số nguyên.")

  danh_sach = [random.randint(1, 100) for _ in range(n)]
  print("-" * 30)

  while True:
    print("\n===== MENU =====")
    print("1. In ra danh sách vừa tạo.")
    print("2. In đảo ngược danh sách.")
    print("3. Sắp xếp danh sách và in ra danh sách vừa sắp xếp.")
    print("4. Tìm phần tử lớn nhất và vị trí cuối cùng của nó.")
    print("5. Đếm số lượng phần tử X và in vị trí xuất hiện.")
    print("6. In ra vị trí các phần tử là số nguyên tố.")
    print("7. Tìm các số duy nhất (không trùng lặp) trong danh sách.")
    print("8. Liệt kê giá trị và số lần xuất hiện của nó.")
    print("9. In ra các đoạn con giảm liên tiếp.")
    print("0. Thoát chương trình.")
    print("================")

    while True:
        try:
            lua_chon_str = input("Nhập lựa chọn của bạn: ")
            lua_chon = int(lua_chon_str)
            if 0 <= lua_chon <= 9:
                break
            else:
                print("Lựa chọn không hợp lệ. Vui lòng chọn từ 0 đến 9.")
        except ValueError:
            print("Giá trị nhập không hợp lệ. Vui lòng nhập một số nguyên.")

    print("-" * 30)

    if lua_chon == 1:
      print("1. Danh sách vừa tạo:")
      print(danh_sach)
    elif lua_chon == 2:
      print("2. Danh sách đảo ngược:")
      print(danh_sach[::-1])
    elif lua_chon == 3:
      print("3. Danh sách đã sắp xếp:")
      danh_sach_sap_xep = sorted(danh_sach)
      print(danh_sach_sap_xep)
    elif lua_chon == 4:
      print("4. Phần tử lớn nhất và vị trí cuối cùng:")
      if not danh_sach:
          print("Danh sách rỗng.")
      else:
          max_val = max(danh_sach)
          last_index = -1
          for i in range(len(danh_sach) - 1, -1, -1):
              if danh_sach[i] == max_val:
                  last_index = i
                  break
          print(f"   - Phần tử lớn nhất: {max_val}")
          print(f"   - Vị trí cuối cùng của phần tử lớn nhất: {last_index}")
    elif lua_chon == 5:
      print("5. Đếm số lượng phần tử X:")
      while True:
        try:
          x_str = input("   Nhập giá trị X cần đếm: ")
          x = int(x_str)
          break
        except ValueError:
          print("   Giá trị nhập không hợp lệ. Vui lòng nhập một số nguyên.")

      count = danh_sach.count(x)
      indices = [i for i, val in enumerate(danh_sach) if val == x]

      print(f"   - Số lượng phần tử bằng {x}: {count}")
      if indices:
        print(f"   - Các vị trí xuất hiện của {x}: {indices}")
      else:
        print(f"   - Phần tử {x} không có trong danh sách.")
    elif lua_chon == 6:
      print("6. Vị trí các phần tử là số nguyên tố:")
      prime_indices = [i for i, val in enumerate(danh_sach) if is_prime(val)]
      if prime_indices:
        print(f"   - Các vị trí của số nguyên tố: {prime_indices}")
        print(f"   - Các số nguyên tố tương ứng: {[danh_sach[i] for i in prime_indices]}")
      else:
        print("   - Không có số nguyên tố nào trong danh sách.")
    elif lua_chon == 7:
      print("7. Các số duy nhất trong danh sách:")
      unique_elements = sorted(list(set(danh_sach)))
      print(unique_elements)
    elif lua_chon == 8:
      print("8. Thống kê tần suất xuất hiện:")
      if not danh_sach:
          print("Danh sách rỗng.")
      else:
          frequencies = Counter(danh_sach)
          print("   Giá trị : Số lần xuất hiện")
          for value in sorted(frequencies.keys()):
              print(f"   - {value:<7} : {frequencies[value]}")
    elif lua_chon == 9:
        print("9. Các đoạn con giảm liên tiếp:")
        decreasing_subs = find_decreasing_subsequences(danh_sach)
        if decreasing_subs:
            for i, sub in enumerate(decreasing_subs):
                print(f"   - Đoạn {i+1}: {sub}")
        else:
            print("   - Không tìm thấy đoạn con nào giảm liên tiếp (có độ dài > 1).")
    elif lua_chon == 0:
      print("Kết thúc chương trình. Tạm biệt!")
      break

    input("\nNhấn Enter để tiếp tục...")
```
